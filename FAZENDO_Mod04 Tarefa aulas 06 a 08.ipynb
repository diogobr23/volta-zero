````python
#1) Seu código aqui

def pesquisa(est:str,tip_inte:str):
  if tip_inte == 'Municipio':
    tip_inte = 'Munic%C3%ADpio'
  url2 = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={est}&tipo_interessado={tip_inte}'
  print(url2)
  r2 = requests.get(url2)
  data_json2 = r2.json()
  return data_json2
````

````python
# 2) Seu código aqui
df_exercicio = pesquisa('MG','Estado')
df_PVL = pd.DataFrame(df_exercicio['items'])
cond = df_PVL['status'] == 'Arquivado por decurso de prazo'
df_PVL[cond]['status'].value_counts()
````

````python
# 3) Seu código aqui
df_exercicio = pesquisa('BA','Municipio')
df_PVL = pd.DataFrame(df_exercicio['items'])
cond = df_PVL['status'] == 'Deferido'
df_PVL[cond]['interessado'].value_counts()
````

````python
# 4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

# utilizar a função para selecionar as informações da Bahia
df_exercicio = pesquisa('BA', 'Estado')

# cração de um novo DataFrame df_Bahia com os dados selecionados
df_Bahia = pd.DataFrame(df_exercicio['items'])

# Utilizar o método .to_csv para criar o arquivo com o DataFrame df_Bahia em um arquivo .csv
df_Bahia.to_csv('solicitacoes_bahia_estado.csv', index=False)
````
